```python
import pandas as pd
dataset = pd.read_csv("50_Startups.csv")
dataset = pd.get_dummies(dataset,drop_first=True,dtype=int)
dataset.columns
```

```python
independent = dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']]
dependent = dataset[['Profit']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
# Standardisation is a defined range
#converting input data i.e x_train to defined range(-1 to 1 or -2 to 2)
#If difference of min and max value of a coulmn in input dataset is large then 
#we do conversion to some defined range (-1 to 1) i.e Standardization.
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
x_train
```

```python
#Model creation using Support Vector Machine
from sklearn.svm import SVR
#regressor = SVR(kernel="linear")
#regressor = SVR(kernel = "poly")
regressor = SVR(kernel='rbf', C=10000)
regressor.fit(x_train, y_train)
             
```

```python
y_predict = regressor.predict(x_test)
#Evaluation Metrics
from sklearn.metrics import r2_score
r_score = r2_score(y_test,y_predict)
r_score

#by increasing c value in SVR, the r_score value is increasing.
```

```python
#Here _rscore value is in negative , so this consider as not a good model for the particular input dataset.
#Lets try with SVR(kernel = "poly") or "sigmoid" or "precomputed" 
```
